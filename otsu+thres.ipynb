```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('2.5bright.jpg', 0)  # load as grayscale image
```

```python
img_blur = cv2.GaussianBlur(img, (5, 5), 0)  # kernel size (5, 5)
```

```python
img_laplacian = cv2.Laplacian(img_blur, cv2.CV_64F)
img_log = cv2.GaussianBlur(img_laplacian, (5, 5), 0)  # kernel size (5, 5)
```

```python
# convert img_log to 8-bit image
img_log_8u = cv2.convertScaleAbs(img_log)

# apply thresholding to the 8-bit image
_, img_thresh = cv2.threshold(img_log_8u, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imwrite('otsu+thres.png',img_thresh)
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(img, cmap='gray'),plt.axis('off'), plt.title('Original Image')
plt.subplot(122), plt.imshow(img_thresh, cmap='gray'),plt.axis('off'), plt.title('Enhanced Image')

plt.show()
```
