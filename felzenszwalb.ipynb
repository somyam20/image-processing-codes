```python
import skimage.segmentation
from matplotlib import pyplot as plt
image=plt.imread("2.5cbrightness.jpg")
seg=skimage.segmentation.felzenszwalb(image, scale=1200)
seg2=skimage.segmentation.felzenszwalb(image, scale=1000)
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(seg); ax1.set_xlabel("k=1000")
ax2.imshow(seg2); ax2.set_xlabel("k=600")

plt.axis('off')
plt.imshow(seg2)
plt.savefig('fel.png')
```
