```python
import cv2
import numpy as np


start_point = (-1, -1)
end_point = (-1, -1)
drawing = False

def mouse_event(event, x, y, flags, param):
    global start_point, end_point, drawing

    if event == cv2.EVENT_LBUTTONDOWN:
        start_point = (x, y)
        drawing = True
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            end_point = (x, y)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False

def main():
    global start_point, end_point  

    
    original_image_path = r"C:\Users\somya\3 H.tif"
    original_image = cv2.imread(original_image_path)
    resized_image = cv2.resize(original_image, (800, 600))

    # Create a window for the resized image
    cv2.namedWindow('Resized Image')
    cv2.setMouseCallback('Resized Image', mouse_event)

    while True:
        image = resized_image.copy()

        if start_point != (-1, -1) and end_point != (-1, -1):
            cv2.rectangle(image, start_point, end_point, (0, 255, 0), 2)

        cv2.imshow('Resized Image', image)

        key = cv2.waitKey(1)

        if key == ord('r'):
            start_point = end_point = (-1, -1)
        elif key == ord('n'):
           
            input_image_path = r"C:\Users\somya\3 H_labelled.tif"
            input_image = cv2.imread(input_image_path)
            resized_input_image = cv2.resize(input_image, (800, 600))

            
            if start_point != (-1, -1) and end_point != (-1, -1):
                cv2.rectangle(resized_input_image, start_point, end_point, (0, 255, 0), 2)

                
                roi_original = original_image[start_point[1]:end_point[1], start_point[0]:end_point[0]]
                cv2.imwrite('roi_original.jpg', roi_original)

                
                roi_input = input_image[start_point[1]:end_point[1], start_point[0]:end_point[0]]
                cv2.imwrite('roi_input.jpg', roi_input)

            
            cv2.imshow('Resized Input Image', resized_input_image)
            cv2.waitKey(0)
            cv2.destroyWindow('Resized Input Image')

        elif key == 27:  
            break

    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
